```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
fname = r"build/gmaps.txt"

f = open(fname,"r")
lines = f.readlines()

for line in lines:
    print(line, end="")
```

```python
LABELS = ("$", "R", "?", "E", "M", "T", "B", "N", "F")
COLORS = ("tab:blue","tab:green","tab:purple","tab:black","tab:red","tab:black","tab:white","tab:pink")
def mapGraph(line):
    arr = [int(elt) for elt in line.split(",")[:210]]
    pos = dict()
    room = dict()
    labels = dict()
    
    roomSets = [set() for elt in LABELS]
    
    D = nx.DiGraph()
    
    # add all vertices first
    i = 0
    for y in range(15):
        for x in range(7):
            lvalue = arr[2*i]
            if lvalue != 7:
                mynode = (y,x)
                D.add_node(mynode)
                pos[mynode] = (y,x)
                labels[mynode] = LABELS[lvalue]
                
                roomSets[lvalue].add(mynode)
            i += 1
    
    
    print(len(D.nodes()), "vertices...")
    print(len(labels), "vertices...")
    
    print(D.nodes())
    
    i = 0
    for y in range(14):
        for x in range(7):
            if arr[2*i] != 7:
                for x2 in range(7):
                    if (1 << x2) & arr[2*i+1]:
                        node2 = (y+1, x2)
                        if node2 in D.nodes():
                            D.add_edge((y,x), node2)
                            print(D.nodes())
        i += 1
    
    print(len(D.nodes()), "vertices...")
    print(len(labels), "vertices...")
    
    BOSS = (15, 3)
    D.add_node(BOSS)
    pos[BOSS] = (15, 3)
    labels[BOSS] = "B"
    roomSets[6].add(BOSS)
    
    #print([len(elt) for elt in roomSets])
    for i in range(9):
        nx.draw_networkx_nodes(D, pos, nodelist=roomSets[i], node_color="tab:red")
    
    #nx.draw_networkx_labels(D, pos, labels)
    return nx.draw_networkx(D, pos)
```
